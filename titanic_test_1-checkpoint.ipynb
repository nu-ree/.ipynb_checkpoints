- 기본 연산하기

```python
train["Age"].min() #가장 어린 탑승자 
```

```python
age_mean=train["Age"].mean()
age_mean
```

```python
train.loc[train["Age"].isnull(), "Age"]= age_mean
#train.loc[train["Age"].isnull(), "Age"]= train["Age"].mean()
train
```

```python
train["DataType"] = "test"
train.head()
```

```python
train["Familysize"] = train["SibSp"]+train["Parch"]+1
train
```

```python
train.drop("DataType", axis=1) # 열 드롭은 axis = 1, 행 드롭은 axis=0
train
```

```python
embarked = pd.get_dummies(train["Embarked"],prefix="Embarked") # one hot encoding 더미로 변환해서 embarked 데이터 만들어주기 
embarked
```

```python
pd.concat([train,embarked], axis=1) #concat = concate 옆으로 붙여주기 
```

```python
train.to_csv("modified_training.csv") #csv로 저장해주기 
```
