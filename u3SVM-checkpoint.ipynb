### 커널 서포트 벡터 머신
- 일반 SVM을 확장하여 입력 데이터를 다차원으로 변경하여 추가로 변수를 만들어 사용하는 방법을 커널 SVM이라고 한다. 분류에 사용하는 커널 SVM을 SVC로 부른다. 회귀분석에서는 SVR을 사용한다.
- 선형모델의 한계를 극복하기 위해서 (더 다양한 변수를 만들어내기 위해서) 변수끼리 곱하거나, 변수의 자승값 등을 새로운 변수로 활용하는 방법이다
- 앞에서 사용한 간단한 데이터 모델을 다시 보자

```python
X, y = make_blobs(centers=4, random_state=8)
y = y % 2

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

```python
from sklearn.svm import LinearSVC
linear_svm = LinearSVC().fit(X, y)

mglearn.plots.plot_2d_separator(linear_svm, X)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

```python
# 두 번째 특성을 제곱하여 추가합니다
X_new = np.hstack([X, X[:, 1:] ** 2])

from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
# 3차원 그래프
ax = Axes3D(figure, elev=-152, azim=-26)
# y == 0 인 포인트를 먼저 그리고 그 다음 y == 1 인 포인트를 그립니다
mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b',
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^',
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

# 선형 결정 경계 그리기
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b',
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^',
           cmap=mglearn.cm2, s=60, edgecolor='k')

ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()],
             cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

### 커널 기법
- 프로그래머가 일일이 고차원 항을 추가하는 방법은 불편하므로 같은 효과를 내도록 자동으로 고차원 항목을 내부적으로 생성하여 모델을 만드는 기법을 커널 기법이라고 한다
- 고차항을 만드는 방법은 크게 두가지인데, 다항식을 지정한 차수까지 추가하는 다항식 커널 방법이 있고, 가우시안 커널기법(Radial Basis Function, RBF) 기법이 있다. 가우시안 커널은 모든 차수의 다항식을 고려한다고 볼 수 있다.
### SVM 이해하기
- SVM에서는 각 샘플 포인트가 결정경계를 구분하는데 얼마나 기여하는지를 측정한다. 당연히 경계 근처에 있는 샘플들이 경계를 나누는데 중요한 역할을 한다. 이러한 샘플들을 서포트 벡터라고 부른다.
- 새로운 데이터에 대해 예측하려면 각 서포트 벡터와의 거리를 계산해봐야 한다.
- 각 서포트 벡터의 중요도는 훈련과정에서 학습한다 (dual*_coef*_ 변수에 저장된다)
- 데이터 포인트 사이의 거리는 가우시안 거리로 계산한다 (p.131 참조)
- 결정경계가 직선이 아닌 곡선을 얻으며, 서포트 벡터들을 표시했다

```python
from sklearn.svm import SVC

X, y = mglearn.tools.make_handcrafted_dataset()                                                                  
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)                                                
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
# 데이터 포인트 그리기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
# 서포트 벡터
sv = svm.support_vectors_
# dual_coef_ 의 부호에 의해 서포트 벡터의 클래스 레이블이 결정됩니다
sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

### SVM 파라미터 튜닝
- 파라미터 감마는, 가우시안 커널 폭의 역수이다. 하나의 훈련 샘플이 미치는 영향의 범위를 결정한다. 작은 값은 넓은 영역을 나타내며 샘플의 영향 범위도 커진다.
- C 값은 규제정도를 나타내며 각 포인트의 중요도를 제한한다
- C를 다르게 설정할 때의 변화는 아래와 같다. C가 작을수록 제약이 크다(과소적합). C가 크면 각 포인트가 모델에 영향을 크게 주어서 과적합이 된다.
- 작은 감마값은 가우시안 커널의 반경을 크게 하여 많은 포인트들이 가까이 있는 것으로 고려 (좌측이 과소적합, 우측이 과적합)
- 

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

for ax, C in zip(axes, [-1, 0, 3]):
    for a, gamma in zip(ax, range(-1, 2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)
        
axes[0, 0].legend(["클래스 0", "클래스 1", "클래스 0 서포트 벡터", "클래스 1 서포트 벡터"],
                  ncol=4, loc=(.9, 1.2))
```

- SVM은 스케일링에 매우 민감하다 (입력의 특성 범위가 비슷해야 한다)

### SVM 을 위한 데이터 전처리
- 여기서는 간단히 0~1 사이로 스케일링해보았다

```python
# 훈련 세트에서 특성별 최소값 계산
min_on_training = X_train.min(axis=0)
# 훈련 세트에서 특성별 (최대값 - 최소값) 범위 계산
range_on_training = (X_train - min_on_training).max(axis=0)

# 훈련 데이터에 최소값을 빼고 범위로 나누면
# 각 특성에 대해 최소값은 0 최대값은 1 임
X_train_scaled = (X_train - min_on_training) / range_on_training
print("특성별 최소 값\n{}".format(X_train_scaled.min(axis=0)))
print("특성별 최대 값\n {}".format(X_train_scaled.max(axis=0)))
```

```python
# 테스트 세트에도 같은 작업을 적용하지만
# 훈련 세트에서 계산한 최소 값과 범위를 사용합니다(자세한 내용은 3장에 있습니다)
X_test_scaled = (X_test - min_on_training) / range_on_training
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("테스트 세트 정확도: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

```python
svc = SVC(C=1000)
svc.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("테스트 세트 정확도: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

- SVC는 다양한 데이터에 대해 잘 동작한다.
- 복잡한 결정경계를 만들 수 있다
- 샘플이 많을 때는 시간이 오래 걸린다
- 최적의 하이퍼 파라미터 설정이 어렵다(C와 감마)
- 이에 비해 랜덤 포레스트가 전처리가 필요없고, 설명이 쉽다
